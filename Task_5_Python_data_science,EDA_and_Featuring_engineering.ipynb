**3.Write a function to cap outliers in all numeric columns.**

```python
# finding the columns in which outliers are present
outliers_columns=[]
for col in column1:
  q1=diabetes[col].quantile(0.25)
  q3=diabetes[col].quantile(0.75)
  IQR=q3-q1
  upper_bound=q3+1.5*IQR
  lower_bound=q1-1.5*IQR
  outliers=diabetes[(diabetes[col]<lower_bound) | (diabetes[col]>upper_bound)]
  num_outliers=len(outliers)
  if num_outliers > 0:
    print("Total outliers present in {} column :".format(col),num_outliers)
    outliers_columns.append('{}'.format(col))
```

```python
outliers_columns
```

```python
#capping all columns in which outliers are present

lower_percentile=0.05
higher_percentile=0.95
for col in outliers_columns:
  print('{}'.format(col),'before capping \n',diabetes[col])
  low,high=diabetes[col].quantile([lower_percentile,higher_percentile])
  diabetes[diabetes[col]<low]=low
  diabetes[diabetes[col]>high]=high
  print(lower_percentile,'low_percentile',low)
  print(higher_percentile,'high_percentile',high)
  print('{}'.format(col),'after capping \n',diabetes[col])
```

```python
#create function to cap outliers
import pandas as pd
import numpy as np

def cap_outliers(diabetes,outliers_columns):
    diabetes_outliers_capped = diabetes.copy()

    for col in outliers_columns:
        q1 = diabetes[col].quantile(0.25)
        q3 = diabetes[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        diabetes_outliers_capped[col] = np.clip(diabetes_outliers_capped[col], lower_bound, upper_bound)

    return diabetes_outliers_capped
```

```python
#calling function cap_outliers
diabetes_capped = cap_outliers(diabetes,outliers_columns)
diabetes_capped
```

**5.Write a function that return dataframe with no multi-colinearity .**


```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def remove_multicollinearity(diabetes, threshold=5.0):
    df_no_multicollinearity = diabetes.copy()
    variables = df_no_multicollinearity.columns

    while True:
        vif = pd.DataFrame()
        vif['Variable'] = variables
        vif['VIF'] = [variance_inflation_factor(df_no_multicollinearity.values, i) for i in range(df_no_multicollinearity.shape[1])]

        max_vif = vif['VIF'].max()
        if max_vif <= threshold:
            break

        max_vif_variable = vif.loc[vif['VIF'].idxmax(), 'Variable']
        df_no_multicollinearity = df_no_multicollinearity.drop(columns=max_vif_variable)
        variables = df_no_multicollinearity.columns

    return df_no_multicollinearity
```
