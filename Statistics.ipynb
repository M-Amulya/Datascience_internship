###**3. Prepare boxplot and count the number of outliers present out of IQR. [Click_Rate]**


```python
import matplotlib.pyplot as plt

# Calculating IQR and setting upper and lower threshold values
q1=df['click_rate'].quantile(0.25)
q3=df['click_rate'].quantile(0.75)
IQR=q3-q1
upper_bound=q3+1.5*IQR
lower_bound=q1-1.5*IQR
outliers=df[(df['click_rate']<lower_bound) | (df['click_rate']>upper_bound)]
num_outliers=len(outliers)
print("number of outliers present out of IQR in click_rate column :",num_outliers)

# Ploting boxplot
plt.figure(figsize=(8,6))
plt.boxplot(df['click_rate'],showfliers=True)
outlier_indices = outliers.index.tolist()
for i in outlier_indices:
    plt.plot(1, df['click_rate'][i], 'ro', markersize=5)

plt.show()

```

###**4. Study the linear relationship among all columns using scatter plot and find out high linearly correlated columns.**


```python
import pandas as pd
import matplotlib.pyplot as plt

# Get the list of numeric column names
columns=df.select_dtypes(include=np.number).columns.tolist()

# Calculate the number of rows and columns for the subplot arrangement
n = len(columns)
rows = min(n,5)
cols = (n + rows - 1) // rows

# Create scatterplot matrix
plt.figure(figsize=(12, 12))
for i in range(n):
    for j in range(n):
      if i * n + j + 1 <= rows * cols:
        plt.subplot(rows, cols, i * n + j + 1)
        plt.scatter(df[columns[j]], df[columns[i]],s=2)
        plt.xlabel(columns[j])
        plt.ylabel(columns[i])

plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix
numeric_cols=df.select_dtypes(include="number")
corr_matrix = numeric_cols.corr(method='pearson')

# Define the correlation threshold value
threshold = 0.7

# Find column pairs with correlation above the threshold
highly_correlated_pairs = []
columns = corr_matrix.columns
for i in range(len(columns)):
  for j in range(i + 1, len(columns)):
    if abs(corr_matrix.iloc[i, j]) > threshold:
      pair = (columns[i], columns[j], corr_matrix.iloc[i, j])
      highly_correlated_pairs.append(pair)

# Print the highly correlated column pairs
for pair in highly_correlated_pairs:
  print(f"Columns '{pair[0]}' and '{pair[1]}' are highly correlated with correlation coefficient {pair[2]}")


```

###**1. X is normally distributed, and the mean is 12 and standard deviation is 4. Find out the probability of the following:**
(i) P( X >= 20 ) (ii) P( X <= 20 ) (iii) P( 0 <= X <= 12)

```python
from scipy.stats import norm
mean = 12
std = 4

# (i) P(X >= 20)
res1 = 1 - norm.cdf(20, loc=mean, scale=std)
print("P(X >= 20):", res1)

# (ii) P(X <= 20)
res2 = norm.cdf(20, loc=mean, scale=std)
print("P(X <= 20):", res2)

# (iii) P(0 <= X <= 12)
res3 = norm.cdf(12, loc=mean, scale=std) - norm.cdf(0, loc=mean, scale=std)
print("P(0 <= X <= 12):", res3)
```

###**2. The mean yield for one-acre plot is 662 kilos with a s.d. 32 kilos. Assuming normal distribution, how many one-acre plots in a batch of 1000 plots would you expect to have yield.**
(i) over 700 kilos (ii) below 650 kilos, and (iii) what is the lowest yield of the best 100
plots?

```python
from scipy.stats import norm

mean = 662
std = 32

# (i) Over 700 kilos P(X>700)
z_score_700 = (700 - mean) / std
Prob_over_700= 1 - norm.cdf(z_score_700)
expected_plots_over_700 = Prob_over_700 * 1000
print("Expected number of plots with yield over 700 kilos:", expected_plots_over_700)

# (ii) below 650 kilos P(X<650)
z_score_650 = (650 - mean) / std
prob_below_650 = norm.cdf(z_score_650)
expected_plots_below_650 = prob_below_650 * 1000
print("Expected number of plots with yield below 650 kilos:", expected_plots_below_650)

# (iii) Lowest yield of the best 100 plots
z_score_best_100 = norm.ppf(0.99)
lowest_yield_best_100 = mean + z_score_best_100 * std
print("Lowest yield of the best 100 plots:", lowest_yield_best_100)
```

###**5. An engineer works for Coca Cola. He knows that the filling machine has a variance of 6 cl. (the filling process can be approximated by a Normal Distribution). Knowing this, he sets the machine to a target fill of 348 cl. In a routine check with 25 cans, he measures an average of 345 cl. Is it possible that the machine is malfunctioning?**

```python
import math
from scipy.stats import t

sample_size = 25
sample_mean = 345
target_fill = 348
variance = 6

# Calculate the standard error
standard_error = math.sqrt(variance / sample_size)

# Calculate the t-value
t_value = (sample_mean - target_fill) / standard_error

# Degrees of freedom for a one-sample t-test
df = sample_size - 1

# Calculate the critical value at a 5% significance level
critical_value = t.ppf(0.05, df)

print("t_value",t_value)
print("critical_value",critical_value)
# Compare the t-value with the critical value
if t_value < -critical_value or t_value > critical_value:
    print("The machine may be malfunctioning.")
else:
    print("The machine is likely functioning within the expected range.")
```
