### **Q1: Create Your own filter and reduce function same as real filter and reduce function?**




Own Filter function







```python
#creating a function which checks if a number is even
def is_even(number):
    return number % 2 == 0

#creating my own filter function
def my_filter(function,iterable):
    return [item for item in iterable if function(item)]

# Example usage of my own filter function

numbers = [1, 2, 3, 4, 5, 6]
# my_filter function using the is_even function and numbers as arguments
result1 = my_filter(is_even, numbers) #calling my_filter function
print(result1)

# Example usage of in_built filter function

result2=list(filter(lambda x:(x%2==0), numbers))
print(result2)   #both gives same output
```

Own Reduce function

```python
def add(x, y):
    return x + y

def my_reduce(function2, iterable, initializer=None):
    iterator = iter(iterable)
    accumulated_value = initializer
    for item in iterator:
        accumulated_value = function2(accumulated_value, item)
    return accumulated_value

# Example usage of my own reduce function

numbers = [1, 2, 3, 4, 5]
# my_reduce function using the add function and numbers as arguments
result1 = my_reduce(add, numbers, initializer=0)
print(result1)

# Example usage of in_built reduce function
from functools import reduce
result2=reduce(lambda x,y:(x+y), numbers)
print(result2)   #both gives same output
```

###**Q5: Create a class which contains all the above functions and import this class for use. (OOPS Concept)**

```python
#creating class1 which contains all above functions
class class1:
    def is_even(self, number):
        return number % 2 == 0

    def my_filter(self, function, iterable):
        return [item for item in iterable if function(item)]

    def add(self, x, y):
        return x + y

    def my_reduce(self, function2, iterable, initializer=None):
        iterator = iter(iterable)
        accumulated_value = next(iterator) if initializer is None else initializer
        for item in iterator:
            accumulated_value = function2(accumulated_value, item)
        return accumulated_value
```

```python
#accessing all methods present in class1
obj=class1()
res1=obj.is_even(6)

iterable=[1,2,3,4,5]
res2=obj.my_filter(obj.is_even,iterable)

res3=obj.add(5,6)

res4=obj.my_reduce(obj.add, iterable)

print(res1)
print(res2)
print(res3)
print(res4)
```

```python
all_mobile_names=[]
all_mobile_prices=[]
Tot_reviews=[]
for i in range (1,21):
  url='https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_5_0_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_5_0_na_na_na&as-pos=5&as-type=HISTORY&suggestionId=mobiles&requestId=1e6397d1-e66a-4360-a542-79816cf77637&page='+str(i)
  req=requests.get(url)
  content=bs(req.content,'html.parser')
  name=content.find_all("div",{"class":"_4rR01T"})
  price=content.find_all("div",{"class":"_30jeq3 _1_WHN1"})
  reviews=content.find_all("div",{"class":"_3LWZlK"})


  for i in name:
    all_mobile_names.append(i.text)
  for i in price:
    all_mobile_prices.append(i.text)
  for i in reviews:
    Tot_reviews.append(i.text)

```

###**Q7: By using selenium. Create a web scrapper for scrapping any product name, product price and review from the first 50 page of flipkart and store all the data into a database like MySQL. Product name will be taken from user's side.**

```python
from selenium import webdriver
import mysql.connector
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time

# Set up MySQL connection
cnx = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="4650",
    database="flip1"
)
cursor = cnx.cursor()

# User input for product name
product_name = input("Enter the product name: ")

# Set up Selenium Chrome driver
driver = webdriver.Chrome()

for page in range(1, 51):
    url = f"https://www.flipkart.com/search?q={product_name}&page={page}"
    driver.get(url)

    product_elements = driver.find_elements(By.XPATH,"//div[@class='_1AtVbE']")
    product_count = len(product_elements)
    print(f"Page {page}: Scraped {product_count} products")

    for product_element in product_elements:
        try:
            name_element = product_element.find_element(By.XPATH,"//div[@class='_4rR01T']")
            product_name = name_element.text[:10]
        except NoSuchElementException:
            pass
        try:
            price_element = product_element.find_element(By.XPATH,"//div[@class='_30jeq3 _1_WHN1']")
            product_price = price_element.text
        except NoSuchElementException:
            pass
        try:
            review_element = product_element.find_element(By.XPATH,"//div[@class='_3LWZlK']")
            product_review = review_element.text
        except NoSuchElementException:
            pass


        # Store data in the database
        insert_query = "INSERT INTO products1 (name, price, review) VALUES (%s, %s, %s)"
        data = (product_name, product_price, product_review)
        cursor.execute(insert_query, data)
        cnx.commit()
    next_button = driver.find_elements(By.XPATH,"//a[@class='_1LKTO3']//span[contains(text(),'Next')]")
    if len(next_button) > 0:
        next_button[0].click()
        time.sleep(2)
    else:
        break

# Close the connections and quit the driver
cursor.close()
cnx.close()
driver.quit()
```
